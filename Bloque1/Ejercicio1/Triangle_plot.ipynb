```python
# Show plots inline, and load main getdist plot module and samples class
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import IPython
import sys, os
path = os.getcwd()
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
%matplotlib inline
```

```python
# Define getdist-subplotter
analysis_settings = {'ignore_rows': '0.1'}
g = plots.get_subplot_plotter(analysis_settings=analysis_settings)

# Get MCMC-samples
samples1 = g.samples_for_root(path + '/datos/base_omegak_w_wa_CMBLens_SN')
samples2 = g.samples_for_root(path + '/datos/base_omegak_w_wa_CMBLens_BAORSD')
samples3 = g.samples_for_root(path + '/datos/base_omegak_w_wa_CMBLens_BAORSD_DES_SN')
```

```python
# Customized triangle plot
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.8
g.settings.title_limit_fontsize = 14
g.triangle_plot([samples1, samples2, samples3], ['omegak', 'w', 'wa'], 
    filled=True, 
    legend_labels=['Planck+Pantheon', 'Planck+SDSS', 'all'], 
    legend_loc='upper right', 
    line_args=[{'lw':2, 'color':'red'},
               {'lw':2, 'color':'blue'},
               {'lw':2, 'color':'gray'}], 
    contour_colors=['red','blue','gray'],
    markers={'omegak':0, 'w':-1, 'wa':0})

g.subplots[0,0].text(-0.029,0.95, "$ow_0w_a$CDM")
```
