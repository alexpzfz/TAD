```python
# Show plots inline, and load main getdist plot module and samples class
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import IPython
import sys, os
path = os.getcwd()
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
%matplotlib inline
```

```python
# Define getdist-subplotter
analysis_settings = {'ignore_rows': '0.1'}
g = plots.get_subplot_plotter(analysis_settings=analysis_settings)

# Get MCMC-samples
samples1 = g.samples_for_root(path + '/chains/SN+BBN')
samples2 = g.samples_for_root(path + '/chains/BAO+BBN')
samples3 = g.samples_for_root(path + '/chains/SN+BAO+BBN')
```

```python
g = plots.get_single_plotter(width_inch=4, ratio=1)
g.plot_2d([samples1, samples2, samples3], 'omegam', 'H0', filled=True)
g.add_legend(['SN', 'BAO', 'BAO+SN'], legend_loc='lower right', colored_text=True);
```
