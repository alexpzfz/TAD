```python
# Show plots inline, and load main getdist plot module and samples class
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import IPython
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
%matplotlib inline
```

# SDSS reanalysis --- BAO-only

```python
# Define getdist-subplotter
analysis_settings = {'ignore_rows': '0.3'}
g = plots.get_subplot_plotter(analysis_settings=analysis_settings)

# Get MCMC-samples
samples1 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/my_likelihood/chains/BAO_SDSS_wCDM_Fig4')
samples2 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/SDSS/base_w_CMB_BAO')
samples3 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/SDSS/base_w_SNonly')
samples4 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/SDSS/base_w_BAOonly')
```

```python
g = plots.get_single_plotter(width_inch=4, ratio=1)
g.plot_2d([samples1, samples2, samples3, samples4], 'omegam', 'w',
          lims=[0,0.7,-1.8,-0.3], filled=[False, True, True, True])
g.add_legend(['reanalyzed BAO', 'CMB T&P', 'SN', 'BAO'], legend_loc='lower left');
g.add_y_marker(-1, lw=2)
```

# SDSS reanalysis --- BAO/RSD

```python
# Define getdist-subplotter
analysis_settings = {'ignore_rows': '0.3'}
g = plots.get_subplot_plotter(analysis_settings=analysis_settings)

# Get MCMC-samples
samples1 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/my_likelihood/chains/SDSS_LCDM_Fig11')
samples2 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/SDSS/base_DES_lenspriors')
samples3 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/SDSS/base_BAORSD_lenspriors')
samples4 = g.samples_for_root('/pscratch/sd/c/crisjagq/1er_Taller_DESI/TAD/Bloque2/demo/official_data/Planck/base_plikHM_TTTEEE_lowl_lowE_lensing')
```

```python
g = plots.get_single_plotter(width_inch=6, ratio=0.7)
g.settings.figure_legend_frame = False
g.plot_2d([samples1, samples2, samples3, samples4], 'omegam', 'sigma8',
          lims=[0.13,0.47,0.56,1.21], filled=[False, True, True, True],
          colors=['k', '#FF4040', '#1874CD', 'gray'])
g.add_legend(['reanalyzed SDSS', 'DES', 'SDSS', 'Planck'], legend_loc='upper right');
g.add_x_bands(0.298, 0.022)
g.add_text("Pantheon", 0.40,0.05)
```

```python
g = plots.get_single_plotter(width_inch=6, ratio=0.7)
g.plot_2d([samples1, samples2, samples3, samples4], 'omegam', 'H0',
          lims=[0.13,0.47,53,78], filled=[False, True, True, True],
          colors=['k', '#FF4040', '#1874CD', 'gray'])
g.add_legend(['reanalyzed SDSS', 'DES', 'SDSS', 'Planck'], legend_loc='lower right');
g.add_x_bands(0.298, 0.022)
g.add_y_bands(74.03, 1.42, color='brown')
g.add_text("Pantheon", 0.40,0.05)
g.add_text("$\Lambda$CDM", 0.05,0.05, fontsize=14)
g.add_text("Distance Ladder", 0.9,0.85)
```
